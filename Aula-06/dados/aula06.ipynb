```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Carregamento das bases de dados (já importadas para o workspace)
base = pd.read_csv('trees.csv')
insect = pd.read_csv('insect.csv')
base2 = pd.read_csv('co2.csv')

```

```python
# Slide 6 - gráfico de barras
# Agrupamento dos dados baseado no atributo 'spray', contando e somando os registros
agrupado = insect.groupby(['spray'])['count'].sum()
# Gráfico de barras
agrupado.plot.bar(color = 'gray')
#Grafico de cores
#agrupado.plot.bar(color = ['blue','yellow','red','green','pink','orange'])
```

```python
#Grafico de linha
# filtro de registros específicos da base de dados (Quebec e Mississipi)
q = base2.loc[base2['Type'] == 'Quebec']
m = base2.loc[base2['Type'] == 'Mississippi']
sns.lineplot(x=q['conc'], y=q['uptake'],errorbar=None).set_title('Quebec: Gráfico de Linha')
plt.show()
```

```python
# Gráfico de linha considerando o volume e o atributo "girth" - sem marker
plt.plot(base.Girth,base.Volume)
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
#Grafico de pizza
#
agrupado.plot.pie(legend = True)
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
#fit_reg linha de tendência
sns.regplot(x=base.Girth,y=base.Volume, data = base, fit_reg = False)
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base2.loc[base2['Type'] == 'Quebec']
m = base2.loc[base2['Type'] == 'Mississippi']
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=q['conc'],y= q['uptake']).set_title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(x=m.conc,y= m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
# Criação do histograma, considerando somente o segundo atributo da base de dados e com x divisões (bins)
# A variável 'h' armazena as faixas de valores de Height
plt.hist(base.iloc[:,1], bins = 6)
plt.title('Árvores')
plt.ylabel('Frequência')
plt.xlabel('Altura')
```

```python
#densidade, método kdeplot para densidade
sns.kdeplot(base.iloc[:,1], color='blue').set(title='Árvores')
```

```python
#densidade e histograma
sns.histplot(base.iloc[:,1], kde=True, bins=9, color='blue').set(title='Árvores')
```

```python
#Slide 16
# Visualização de um boxplot
sns.boxplot(data=base.Volume, orient='v').set_title('Árvores')
# Visualização de vários boxplots na mesma imagem
#sns.boxplot(data = base)
```

```python
# mapa de calor
sns.heatmap(base, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
```

```python
# gráfico treemap
fig = px.treemap(insect, path=["spray"], values="count",
                 color_continuous_scale="Viridis", title="Treemap")

# Exibindo o Treemap
fig.show()
```
