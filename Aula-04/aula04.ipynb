```python

#Aula 4 - Estatística Aplicada

# Importação das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

```

```python
##### SIMULAR O PASSEIO ALEATÓRIO EM 2D


num_passos = 1000 # Definir o número de passos


passos_x = np.random.choice([-1, 1], size=num_passos)  # Movimento horizontal
passos_y = np.random.choice([-1, 1], size=num_passos)  # Movimento vertical

trajetoria_x = np.cumsum(passos_x)
trajetoria_y = np.cumsum(passos_y)

# Plotando o passeio gerado:
plt.plot(trajetoria_x, trajetoria_y, linestyle='-', marker='', alpha=0.7, label="Trajetória")


plt.scatter(0, 0, color='red', label="Início")  # Inserindo os marcadores com ponto inicial e final
plt.scatter(trajetoria_x[-1], trajetoria_y[-1], color='blue', label="Fim")  

# Ajustando as configurações do gráfico para melhor visualização
plt.xlabel("Posição X")
plt.ylabel("Posição Y")
plt.title("Passeio Aleatório 2D")
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Linha em y=0
plt.axvline(0, color='black', linestyle='--', linewidth=1)  # Linha em x=0
plt.legend()
plt.grid()
plt.show()
```

```python
##### SIMULAR O MODELO DO PASSEIO ALEATÓRIO APLICADO EM UMA AÇÃO

num_dias = 252  # Um ano de pregão (~252 dias úteis)
preco_inicial = 100  # Preço inicial da ação
retorno_medio = 0.0005  # Retorno médio diário (~0.05%)
volatilidade = 0.02  # Volatilidade diária (~2%)

# Gerar variações diárias 
variação_diaria = np.random.normal(retorno_medio, volatilidade, num_dias)
precos = preco_inicial * np.cumprod(1 + variação_diaria)  # Acumular preços

# Plotar a simulação
plt.plot(precos, label="Simulação do Preço da Ação")
plt.xlabel("Dias")
plt.ylabel("Preço")
plt.title("Simulação de Preço de Ação via Passeio Aleatório")
plt.axhline(preco_inicial, color='red', linestyle='--', label="Preço Inicial")
plt.legend()
plt.grid()
plt.show()
```

```python
##### SIMULAR DADOS ALEATÓRIOS - SLIDE 10    (gerar varias vezes)

# Gerar dados aleatórios com média 50 e desvio padrão 10
dados = np.random.normal(loc=50, scale=10, size=1000)

# Criar o histograma
plt.hist(dados, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(np.mean(dados), color='red', linestyle='dashed', linewidth=2, label=f'Média = {np.mean(dados):.2f}')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Histograma de Dados Aleatórios (Média = 50)')
plt.legend()


# Exibir o gráfico
plt.show()
```
